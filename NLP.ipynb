## Removing Hashtags

```python
# tweets_data['text_clean'] = tweets_data['text_clean'].apply(
tweets_data['text_clean'] = tweets_data['text'].apply(
    lambda x: re.sub(
        r'#(\w+)', r'\1', x
        )
    )

tweets_data.head()
```

## Removing Mention Symbols

```python
tweets_data['text_clean'] = tweets_data['text_clean'].apply(
    lambda x: re.sub(
        r'@\w+', '', x
        )
    )

tweets_data.head()
```

## Removing the URLs

```python
tweets_data['text_clean'] = tweets_data['text_clean'].apply(
    lambda x: re.sub(
        r'http\S+|www.\S+', '', x
        )
    )

tweets_data.head()
```

## Removing Digits from the Tweets

```python

# Remove digits
tweets_data['text_clean'] = tweets_data['text_clean'].apply(
    lambda x: re.sub(
        r'\d+', '', x
        )
    )

tweets_data.head()
```

## Removing Punctutation

```python
# Remove punctuation except hyphens
tweets_data['text_clean'] = tweets_data['text_clean'].apply(
    lambda x: re.sub(
        r'[^\w\s-]', '', x
        )
    )

tweets_data.head()
```

## Removing words like 'golden', 'globe', etc

```python
# Remove specific phrases like "golden globes"
tweets_data['text_clean'] = tweets_data['text_clean'].apply(
    lambda x: re.sub(
        r'\b(golden\s?globes?)\b', '', x, flags=re.IGNORECASE
        )
    )

tweets_data.head()
```

## Removing the stopwords

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    words = text.split()  # Split the text into a list of words
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return " ".join(filtered_words)  # Join the filtered words back into a string

tweets_data['text_without_stopwords'] = tweets_data['text_clean'].apply(remove_stopwords)
tweets_data.head(25)
```

## Tokenizing the tweets for lemmatization and further process

```python
tokenizer = TweetTokenizer()
tweets_data['tokens'] = tweets_data['text_without_stopwords'].apply(
    lambda x: tokenizer.tokenize(x)
    )

tweets_data.head()
```

## Lemmatizing the text to the base word

```python
lemmatizer = WordNetLemmatizer()
tweets_data['lemmatized'] = tweets_data['tokens'].apply(
    lambda tokens: [lemmatizer.lemmatize(token, pos='v') for token in tokens]
    )

tweets_data.head()
```

```python
import re

def extract_award_info(text):
    awards = re.findall(
        r'\b(Best|Outstanding|Top|WINNER)\s+(Actress|Actor|Director|Picture|Performance|Song|Screenplay|Series)\s+(?:in\s+(?:a\s)?)?(Motion Picture|Television Series|Miniseries|Drama|Musical or Comedy|Comedy|Drama)?(?:\s+–\s+|\s+or\s+|\s*,\s*)(Musical or Comedy|Drama|Comedy|Musical|Television|Film)?',
        text,
        re.IGNORECASE
    )
    # Normalize the extracted components
    normalized_awards = []
    for award in awards:
        # Convert all components to lowercase and strip extra spaces
        award_type = award[0].strip().lower()
        category = award[1].strip().lower()
        medium = award[2].strip().lower() if award[2] else ''
        genre = award[3].strip().lower() if award[3] else ''
        temp = str(award_type) + " " + str(category) + " " + str(medium) + " "  + str(genre)
        # Construct a normalized string without commas
        normalized_award = ''.join(filter(None, temp))
        normalized_awards.append(normalized_award)

    return normalized_awards

# Apply the function to the 'text' column
tweets_data['award_info'] = tweets_data['text'].apply(extract_award_info)

tweets_data[tweets_data['award_info'].apply(lambda x: len(x) > 0)]
```

# Finding Nominees

```python
import re
from collections import Counter
import json
import spacy


# Initialize a dictionary to hold the results
awards_nominees = {}

# Process each unique award found
for _, row in tweets_data.iterrows():
    for award in row['award_info']:
        # Standardize award names to lowercase
        award = award.lower().strip()

        # Initialize the award entry if not already present
        if award not in awards_nominees:
            awards_nominees[award] = {'tweets': [], 'nominees': [], 'winners': []}

        # Add the tweet text to the award's tweet list
        row_text = ' '.join(row['lemmatized'])
        awards_nominees[award]['tweets'].append(row_text)

        # Extract potential nominees from the tweet using regex
        nominee_candidates = re.findall(r'\b[A-Z][a-z]*\s[A-Z][a-z]*\b', row_text)

        # Filter out candidates that are part of the award name
        filtered_nominees = [nominee for nominee in nominee_candidates if award not in nominee.lower()]

        # # Validate nominees using the NER "entities" column to ensure they are recognized as PERSON entities
        valid_nominees = set(ent[0] for ent in row['entities'] if ent[1] == 'PERSON')
        final_nominees = [nominee for nominee in filtered_nominees if nominee in valid_nominees]


        # #trial
        # final_nominees = set(ent[0] for ent in row['entities'] if ent[1] == 'PERSON')

        # Add validated nominees to the award entry
        awards_nominees[award]['nominees'].extend(final_nominees)


# Sort the awards by the number of tweets
sorted_awards = sorted(awards_nominees.items(), key=lambda x: len(x[1]['tweets']), reverse=True)

# Display the sorted data and top 5 nominees in a readable form
for award, data in sorted_awards:
    print(f"Award: {award.title()}")
    print(f"Total Tweets: {len(data['tweets'])}")
    print("Top 5 Nominees:")

    # Count and display the top 5 nominees
    nominee_counts = Counter(data['nominees'])
    for nominee, count in nominee_counts.most_common():
        print(f"  - {nominee}: {count} mentions")
    print("\n" + "-"*40 + "\n")

# Save the results to a JSON file
with open('awards_nominees.json', 'w') as f:
    json.dump(awards_nominees, f, indent=4)
```

## Extracting Winners

```python
import re
from collections import Counter
import json
import spacy



for _, row in tweets_data.iterrows():


    for award in row['award_info']:
        # Standardize award names to lowercase
        award = award.lower().strip()

        # Combine the text of the tweet for analysis
        row_text = ' '.join(row['lemmatized'])

        # Check for "win" related words and extract potential winners
        if re.search(r'\b(wins?|won|winning|winner)\b', row_text, re.IGNORECASE):
            # Add the tweet text to the award's tweet list

            # Extract potential winners from the tweet using regex
            winner_candidates = re.findall(r'\b[A-Z][a-z]*\s[A-Z][a-z]*\b', row_text)

            # Filter out candidates that are part of the award name
            filtered_winners = [winner for winner in winner_candidates if award not in winner.lower()]

            # Validate winners using the NER "entities" column to ensure they are recognized as PERSON entities
            valid_winners = set(ent[0] for ent in row['entities'] if ent[1] == 'PERSON')
            final_winners = [winner for winner in filtered_winners if winner in valid_winners and winner in awards_nominees[award]['nominees']]

            # Add validated winners to the award entry
            awards_nominees[award]['winners'].extend(final_winners)

# Sort the awards by the number of tweets
sorted_awards = sorted(awards_nominees.items(), key=lambda x: len(x[1]['tweets']), reverse=True)

# Display the sorted data and top 5 nominees in a readable form
for award, data in sorted_awards:
    if (len(data['winners'])>0):
        print(f"Award: {award.title()}")
        print(f"Total Tweets: {len(data['tweets'])}")
        print("Top 5 Nominees:")

        # Count and display the top 5 nominees
        nominee_counts = Counter(data['nominees'])
        for nominee, count in nominee_counts.most_common(5):
            print(f"  - {nominee}")

        print("Winners Mentioned:")
        winner_counts = Counter(data['winners'])
        for winner, count in winner_counts.most_common(1):
            print(f"  - {winner}")
        print("\n" + "-"*40 + "\n")

# Save the results to a JSON file
with open('awards_winners.json', 'w') as f:
    json.dump(awards_nominees, f, indent=4)
```

## To Find Hosts

```python
from collections import Counter
import re

def extract_names_regex(text):
    pattern = re.compile(r'\b[A-Z][a-z]*\b')
    return pattern.findall(text)

tweets_data['names'] = tweets_data['text'].apply(extract_names_regex)

all_names = [name for names_list in tweets_data['names'] for name in names_list]

name_counts = Counter(all_names)
print(name_counts.most_common(10))
```

```python
def check_host_context(text, names):
    host_related = re.findall(r'\bhost(s|ing)?\b', text.lower())
    if host_related:
        return [name for name in names if name in text]
    return []

tweets_data['potential_hosts'] = tweets_data.apply(lambda row: check_host_context(row['text'], row['names']), axis=1)
potential_hosts = [host for hosts_list in tweets_data['potential_hosts'] for host in hosts_list]
host_counts = Counter(potential_hosts)
most_likely_hosts = host_counts.most_common(10)
print(most_likely_hosts)
```

```python
from itertools import combinations

most_likely_host_names = [host[0] for host in most_likely_hosts]
host_pairs = list(combinations(most_likely_host_names, 2))


def count_pairs_in_text(text, pairs):
    pair_counts = Counter()
    for pair in pairs:
        pair_phrase = ' '.join(pair).lower()
        if pair_phrase in text.lower():
            pair_counts[pair_phrase] += 1
    return pair_counts

all_pair_counts = Counter()
for text in tweets_data['text']:
    pair_counts = count_pairs_in_text(text, host_pairs)
    all_pair_counts.update(pair_counts)

top_5_pairs = all_pair_counts.most_common(2)

host_names = [pair[0].title() for pair in top_5_pairs]
print(f"The hosts are {host_names[0]} and {host_names[1]}.")
```
